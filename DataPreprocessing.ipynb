```python
#war 반영 선수데이터
import pandas as pd
import numpy as np
month = [5,6,7,8,9,10]
day = range(1,32)
stadium = ['부산사직', '고척돔', '잠실', '케이티위즈파크', '창원NC파크', '대전한밭', '챔피언스필드', '라이온즈파크', '인천문학','마산']
year = [2016,2017,2018,2019] #WAR데이터는 2016부터

games = pd.DataFrame()
for y in year:
    #진행상황을 확인하기 위한 print
    print(y)
    for m in month:
        name2 = './WAR/{}0{}.csv'.format(y, m - 1) #직전 달의 WAR데이터 가져오기
        war = pd.read_csv(name2, encoding='cp949', header=None)
        war_dict = {}
        for i in war[1]:
            war_dict[i] = str(war[war[1] == i][5]).split()[1]
        #dict 키에 np.nan이 입력되는 경우 삭제
        try:
            del war_dict[np.nan]
        except:
            print('no nan')
        for i in war_dict:
            war_dict[i] = float(war_dict[i])
    
        for d in day:
            for s in stadium:
                name = './선발/{}-{}-{}_{}_away.csv'.format(y,m,d,s)
                name1 = './선발/{}-{}-{}_{}_home.csv'.format(y,m,d,s)
                # 경기가 열리지 않은 날자의 경우 continue
                try:
                    temp = pd.read_csv(name, encoding='cp949', header=None)
                    temp1 = pd.read_csv(name1, encoding='cp949', header=None)
                except:
                    continue
                #우천취소등으로 선발데이터가 불완전한 경우 continue(데이터 확인 위한 print)    
                try:
                    temp.columns = ['타순', '선수', '포지션', '투타']
                    temp = temp[['선수', '타순', '포지션', '투타']]
                    temp1.columns = ['타순', '선수', '포지션', '투타']
                    temp1 = temp[['선수', '타순', '포지션', '투타']]
                except:
                    print(y,m,d,s)
                    continue
                # 선수이름을 WAR로 수치로 변경 직전달 WAR점수가 없는 신인선수등은 0점 처리 
                
                # WAR이 0이하여도 반영
                temp['선수'] = temp.선수.apply(lambda x: war_dict[x.strip()] if x.strip() in war_dict.keys()   else 0)
                temp1['선수'] = temp1.선수.apply(lambda x: war_dict[x.strip()] if x.strip() in war_dict.keys() else 0)
                
                #WAR이 0이하일 경우 0처리
                #temp['선수'] = temp.선수.apply(lambda x: war_dict[x.strip()] if x.strip() in war_dict.keys() and war_dict[x.strip()] >0 else 0)
                #temp1['선수'] = temp1.선수.apply(lambda x: war_dict[x.strip()] if x.strip() in war_dict.keys() and war_dict[x.strip()] >0 else 0)
                
                game = {}
                game['경기'] = '{}-{}-{}_{}'.format(y,m,d,s)
                #away팀 수비포지션
                for i in temp.포지션:
                    if '지명' in i:
                        continue
                    game['away_' + i] = str(temp[temp.포지션 == i].선수).split()[1].split('\n')[0]
                    
                #away팀 타순   
                for i in range(1, 10):
                    game['away_' + str(i) + '번타자'] = str(temp[temp.타순 == i].선수).split()[1].split('\n')[0]
                #home팀 수비포지션   
                for i in temp1.포지션:
                    if '지명' in i:
                        continue
                    game['home_' + i] = str(temp1[temp1.포지션 == i].선수).split()[1].split('\n')[0]
                #home팀 타순   
                for i in range(1, 10):
                    game['home_' + str(i) + '번타자'] = str(temp1[temp1.타순 == i].선수).split()[1].split('\n')[0]
                   
                
                foo = pd.DataFrame([game])
                games = pd.concat([games, foo]) #games DataFrame에 각 for문의 결과 foo를 concat
print(games)
games.to_csv('game_war.csv', encoding='cp949')
```
