```python
#선발선수데이터
import requests
from SearchList import getSearchList
date = ['201{}-{}-{}'.format(y,m,d) for y in [4,5,6,7,8,9] for m in [3,4,5,6,7,8,9,10] for d in range(1,32)]
stadium = ['부산사직', '고척돔', '잠실', '케이티위즈파크', '창원NC파크', '대전한밭', '챔피언스필드', '라이온즈파크', '인천문학','마산','대구', '사직', '문학', '목동', '울산', '대전', '포항', '광주']
path = ""
away_tag = "body > div.wrapper > div.content-wrapper > div > section.content > div > div:nth-child(2) > div:nth-child(1) > div:nth-child(1) > div > div.box-body.no-padding.table-responsive tr"
home_tag = "body > div.wrapper > div.content-wrapper > div > section.content > div > div:nth-child(2) > div:nth-child(1) > div:nth-child(2) > div > div.box-body.no-padding.table-responsive tr"

for d in date:
    print(d)
    for s in stadium:
        path = "http://www.statiz.co.kr/boxscore.php?opt=5&date={}&stadium={}".format(d,s) #try-catch
        away_search_list = getSearchList(path, away_tag)
        home_search_list = getSearchList(path,home_tag)
        away_list = [x.find_all('td') for x in away_search_list]
        home_list = [x.find_all('td') for x in home_search_list]
        if len(away_search_list) != 0:
            with open('./선발/{}_{}_{}.csv'.format(d,s,'away'), 'w') as f:
                for i in away_list:
                    line = ''
                    for j in i:
                        value = j.text.strip()

                        if '좌타' not in value and '우타' not in value:
                            value = value if value != '' else '0'
                            line += value+', '
                        else:
                            line += value+'\n'
                    f.write(line)
            with open('./선발/{}_{}_{}.csv'.format(d, s, 'home'), 'w') as f:
                for i in home_list:
                    line = ''
                    for j in i:
                        value = j.text.strip()
                        if '좌타' not in value and '우타' not in value:
                            value = value if value != '' else '0'
                            line += value + ', '
                        else:
                            line += value + '\n'
                    f.write(line)
```

# WPA Crwaling 2014-2019

```python
# WPA 긁어오기
import requests
from SearchList import getSearchList

#si_it 4:7, 5:8, 6:9, 7:10, 8:11, 9:12, 10:13
path_form = 'http://www.statiz.co.kr/stat.php?mid=stat&re=0&ys={}&ye={}&se=0&te=&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=16&o1=WPA&o2=AVG&de=1&lr=0&tr=&cv=0&ml=1&sn=30&pa={}&cn=&si=999&si_it={}&si_wd=&si_tm=&si_ha=&si_te=&si_st=&si_as=&si_or=&si_ty=&si_pl=&si_in=&si_on=&si_um=&si_oc=&si_bs=&si_sc=&si_cnt=&si_aft=&si_li='
path = ""
tag = "#mytable  td:nth-child(4)"
tag2 = "#mytable  td:nth-child(2) "
# tag = "#fixcol > table > tbody td:nth-child(4) > font > span"

year = [2014,2015,2016,2017,2018,2019]
month = [4,5,6,7,8,9,10]
pa = [0,30,60,90,120,150,180,210]
for y in year:
    for m in month:
        with open('./WPA/{}-{}WPA.csv'.format(y,m), 'w') as f:
            for p in pa:
                si_it = m+3
                path = path_form.format(y,y,p,si_it)
                WPA_search_list = getSearchList(path, tag)
                name_search_list = getSearchList(path, tag2)

                for name,wpa in zip(name_search_list,WPA_search_list):
                    f.write(name.text+', '+wpa.text+'\n')
```
